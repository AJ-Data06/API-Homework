```python
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import linregress
from citipy import citipy
from numpy import random
import random
from pprint import pprint
```

## Generate Cities List

```python
cities_list = []

#lats = random.sample((lat_range), 500)
#lngs = random.sample((lng_range), 500)
#lats = random.uniform(lat_range, 600)
#lats = random.uniform(lng_range, 600)
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)

lats_lngs = zip(lats,lngs)


for lat_lng in lats_lngs:
    lats_lngs = []
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if lat_lng[0] not in lats_lngs or lat_lng[1] not in lats_lngs:
        lats_lngs.append(lat_lng[0])
        lats_lngs.append(lat_lng[1])
    if city not in cities_list:
        cities_list.append(city)
       
    
    
print(len(cities_list))

    

    

        
        
    


```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
api_key = "375a053fdb55254e2a97c41c9e383900"
params = {"appid": api_key}
base_url = "http://api.openweathermap.org/data/2.5/weather?"
#api.openweathermap.org/data/2.5/weather?q={city name}&appid={API key}
count = 0
index = 1


Country = []
Date = []     
Latitude = []
Longitude = []
Cloudiness = []
Humidity = []
Max_Temp = []
Wind_Speed = []
missing_city = []
cities_list_02 = []
for city in cities_list:
    try:
        params["q"] = city
        count+= 1
        
        response = requests.get(base_url, params=params).json()
        print(f"Processing record {count} for set: {index} city {city}")
        
        
        Country.append(response["sys"]["country"])
        Date.append(response["dt"])
        Latitude.append(response["coord"]["lat"])
        Longitude.append(response["coord"]["lon"])
        Cloudiness.append(response["clouds"]["all"])
        Humidity.append(response["main"]["humidity"])
        Max_Temp.append(response["main"]["temp_max"])
        Wind_Speed.append(response["wind"]["speed"])
        cities_list_02.append(city)
        
        if count == 50:
            index = index + 1
            
    except (KeyError, IndexError):
        
        print(f"Missing city, Skipping: {city}")
        missing_city.append(city)
    
        
        
        
      
         
        
     
        
        
print("------end-----")
    
        
```

## Linear Regression

```python
northern_Hemisphere = clean_df.loc[clean_df["Latitude"] >= 0]
northern_Hemisphere
```

```python
southern_Hemisphere = clean_df.loc[clean_df["Latitude"] <= 0]
southern_Hemisphere
```

Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_value = northern_Hemisphere["Latitude"]
y_value = northern_Hemisphere["Temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_value, y_value)
regress_value = x_value * slope + intercept
equation = "y =" + str(round(slope,2)) + "x +" + str(round(intercept,2))
r_square = f"r^2 = {round(rvalue ** 2,2)}"
plt.scatter(northern_Hemisphere["Latitude"], northern_Hemisphere["Temperature"])
plt.plot(x_value, regress_value, "r-")
plt.annotate(equation, (10,20), fontsize=15, color="red")
plt.annotate(r_square, (10,0), fontsize=15, color="red")
plt.xlabel("Latitude")
plt.ylabel("Temp_(F)")
plt.savefig("Max Temp vs. Latitude Linear Regression.png")
plt.show()
print(f"In Northen Hemisphere, temperature starts to fall when starting to move away from the equator and this regression equation explains {77} percent of the dependent variable is explained by the independent variable")
```
